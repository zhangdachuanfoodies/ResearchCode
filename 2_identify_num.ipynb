```python
from numpy import *
import operator
import matplotlib
import matplotlib.pyplot as plt

def img2vector(filename):
    returnVect = zeros((1, 1024))
    fr = open(filename)
    for i in range(32):
        lineStr = fr.readline()
        for j in range(32):
            returnVect[0, 32*i+j] = int(lineStr[j])
    return returnVect
```

```python
def classify0(inX, dataSet, labels, k):
    """
    inX: inputed vector
    dataSet: training example
    labels: label vector
    k: number of neighbor
    """
    dataSetSize = dataSet.shape[0]
    diffMat = tile(inX, (dataSetSize, 1)) -dataSet  # 把输入向量扩大dataSet倍行数，列数为1
    sqDiffMat = diffMat ** 2
    sqDistances = sqDiffMat.sum(axis=1)  # sum()普通相加  .sum(axis=1)矩阵的每一行向量相加
    distances = sqDistances ** 0.5
    sortedDistIndicies = distances.argsort()  # 返回矩阵中每个元素的排序序号-->matrix([[2, 0, 1]])
    
    classCount = {}
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel, 0) + 1
    sortedClassCount = sorted(classCount.items(), key=operator.itemgetter(1), 
                             reverse=True)
#     print(sortedClassCount)
#     print(sortedClassCount[0][0])
    return sortedClassCount[0][0]
```

```python
def handwritingClassTest():
    from os import listdir   # 列出给定目录的文件名
    hwLabels = []
    trainingFileList = listdir('./example_data/Ch02/trainingDigits')  # 获取目录内容
    m = len(trainingFileList)
    trainingMat = zeros((m, 1024))
    for i in range(m):
        fileNameStr = trainingFileList[i]
        fileStr = fileNameStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        hwLabels.append(classNumStr)
        trainingMat[i, :] = img2vector('./example_data/Ch02/trainingDigits/{}'.format(fileNameStr))
        
        
    testFileList = listdir('./example_data/Ch02/testDigits')
    errorCount = 0.0
    mTest = len(testFileList)
    for i in range(mTest):
        fileNameStr = testFileList[i]
        fileStr = fileNameStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        vectorUnderTest = img2vector('./example_data/Ch02/testDigits/{}'.format(fileNameStr))
        classifierResult = classify0(vectorUnderTest, trainingMat, hwLabels, 2)
        print("the classifier came back with:{:2d}, the real answer is: {:2d}".format(classifierResult, classNumStr))
        if (classifierResult != classNumStr):
            errorCount += 1.0
    print("the total number of errors is: {:.6f}".format(errorCount))
    print("the total error rate is: {:.6f}".format(errorCount/float(mTest)))

handwritingClassTest()
```
